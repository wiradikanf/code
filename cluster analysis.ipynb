Cluster analysis

Peercobaan

```python
import pandas as pd
from sklearn.cluster import KMeans

dataset = pd.read_csv('C:\Hagan\Matakuliah\Semester 3\Mesin Pembelajaran/Project/9/ruspini.csv')
data = dataset.loc[:,['X', 'Y']]
clustering = KMeans(n_clusters=4, init="random", n_init=1)
clusters=clustering.fit_predict(data)

print('\nHasil clustering:\n', clusters)
print('\nSSE = :\n', clustering.inertia_)
```

```python
dataset_test = pd.read_csv('C:\Hagan\Matakuliah\Semester 3\Mesin Pembelajaran/Project/9/test.csv')
dataset_test
```

5 fitur

```python
df_selected_features_5 = dataset_test[['battery_power',	'mobile_wt',	'px_height',	'px_width',	'ram']]
df_selected_features_5
```

```python
test_data = np.array(df_selected_features_5)
newmin = 0
newmax = 1
mindata = test_data.min()
maxdata = test_data.max()
test_data_5 = ((test_data - mindata)*(newmax - newmin)/(maxdata - mindata)) + newmin
test_data_5 = pd.DataFrame(test_data_5)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
inertia = []
for i in range(10):
    clustering = KMeans(n_clusters=3, init="random", n_init=1)
    clusters=clustering.fit_predict(test_data_5)
    inertia.append(clustering.inertia_)

inertia_selected_5 = min(inertia)
print('\nHasil clustering:\n', clusters)
print('\nSSE = :\n', inertia_selected_5)
    
```
