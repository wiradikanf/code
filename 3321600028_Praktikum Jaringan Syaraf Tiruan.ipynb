# Single Perceptron

### AND.csv

```python
import pandas as pd
from sklearn.linear_model import Perceptron
dataset = pd.read_csv('AND.csv')
dataset
```

```python
data = dataset.iloc[:,0:-1]
label = dataset.iloc[:,-1]
clf = Perceptron(tol=1e-3, random_state=0)
clf.fit(data, label)
print(clf.score(data,label))
print(clf.predict(data))
```

# Multi Layer Perceptron

```python
import pandas as pd
from sklearn.neural_network import MLPClassifier as MLP

dataset4 = pd.read_csv('XOR.csv')
dataset4
```

```python
data = dataset4.iloc[:,0:-1]
label = dataset4.iloc[:,-1]
model = MLP(hidden_layer_sizes=(4), max_iter=100, activation = 'relu', learning_rate_init=0.1, solver='sgd')
model.fit(data,label)
print('score:', model.score(data,label))
print('predictions:', model.predict(data))
print('expected:',label)
```

### split untuk data heart.csv dengan metode Hold-out

```python
from sklearn.model_selection import train_test_split

dataset5 = pd.read_csv('heart.csv')
data = dataset5[['thal', 'cp', 'slope']]
datalabel = dataset5.loc[:,['target']]
x_train, x_test, y_train, y_test = train_test_split (dataset5, datalabel, test_size = 0.30, random_state=100)
```

### Dengan multilayer Perceptron lakukan training pada data training dan pilih parameter yang tepat hingga error paling minimal

```python
data = data
label = datalabel
model = MLP(hidden_layer_sizes=(4), max_iter=100, activation = 'relu', learning_rate_init=0.1, solver='sgd')
model.fit(data,label)

print('score:', model.score(data,label))
print('predictions:', model.predict(data))
print('expected:', label)
```

### Lakukan testing untuk data testing dan tampilkan akurasinya

```python
import numpy as np

train_data = np.array(x_train)
train_label = np.array(y_train)
test_data = np.array(x_test)
test_label = np.array(y_test)

klasifikasi = MLP()
klasifikasi.fit(train_data, train_label)
y_pred = klasifikasi.predict(test_data)

from sklearn.metrics import accuracy_score
akurasi = accuracy_score(y_test, y_pred)

print("Tingkat akurasi metode MLP : ", akurasi)
```
