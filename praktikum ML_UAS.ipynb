# 1. dataset

```python
import pandas as pd
dataset = pd.read_csv('e:/data/water_potability.csv')
dataset
```

# 2. data  ambil dataset kolom fitur (Hardness, Solids, Sulfate, Condictivity,Organic_carbon,Potability).


```python
data=dataset[['Hardness','Solids','Sulfate','Conductivity','Organic_carbon','Potability']]
data
```

# 3. train_data  ambil fitur (Hardness, Solids, Condictivity,Organic_carbon,Potability) pada data yang Sulfatenull


```python
train = data[data['Sulfate'].notnull()]
train_data = train[['Hardness','Solids','Conductivity','Organic_carbon','Potability']]
train_data
```

# 4. train_label  ambil fitur (Sulfate) pada data yang Sulfatenull


```python
train_label = train['Sulfate']
train_label
```

# 5. test_data  ambil fitur (Hardness, Solids, Condictivity,Organic_carbon,Potability) pada data yang Sulfate=null

```python
test = data[data['Sulfate'].isna()]
test_data = test[['Hardness','Solids','Conductivity','Organic_carbon','Potability']]
test_data
```

# normalisasi data

## 6. train_data  lakukan normalisasi pada train_data dengan Min-Max 0-1 (catat nilai min dan max setiap atribut)

```python
import numpy as np
dffeat1 = np.array(train_data)
newmin=0
newmax=1
mindata=dffeat1.min()
maxdata=dffeat1.max()
dffeat1=((dffeat1-mindata)*(newmax-newmin)/(maxdata-mindata))+newmin
```

```python
normal_data_train = pd.DataFrame(dffeat1)
normal_data_train.columns = ['Hardness','Solids','Conductivity','Organic_carbon','Potability']
normal_data_train
```

## 7. test_data  lakukan normalisasi pada test_data dengan Min-Max 0-1 (dengan nilai min dan max setiap atribut pada Langkah 6)

```python
dffeat2 = np.array(test_data)
newmin=0
newmax=1
mindata=dffeat2.min()
maxdata=dffeat2.max()
dffeat2=((dffeat2-mindata)*(newmax-newmin)/(maxdata-mindata))+newmin
```

```python
normal_data_test = pd.DataFrame(dffeat2)
normal_data_test.columns = ['Hardness','Solids','Conductivity','Organic_carbon','Potability']
normal_data_test
```

# 8. class_result  Lakukan klasifikasi test_data terhadap train_data dengan 3-NN (output mepakai class pada train_label)

```python
from sklearn import preprocessing
lab = preprocessing.LabelEncoder()
y_transformed = lab.fit_transform(train_label)
y_transformed
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN=KNeighborsClassifier(n_neighbors=3,weights='distance')
KNN.fit(normal_data_train,y_transformed)
prediction = KNN.predict(normal_data_test)
prediction
```

# 9. data (Sulfate)  lakukan pengisian missing values pada data yang Sulfate=null dengan nilai class_result

```python
data.loc[data['Sulfate'].isnull(), 'Sulfate'] = prediction
data
```
