```python
import pandas as pd

survival_counts = dataset.groupby('Pclass')['Survived'].sum()
death_counts = dataset.groupby('Pclass')['Survived'].count() - survival_counts

passenger_counts = pd.concat([survival_counts, death_counts], axis=1)
passenger_counts.columns = ['Survived', 'Not Survived']
passenger_counts
```

```python
import pandas as pd
import seaborn as sns

# Load the data
data = sns.load_dataset('titanic')

# Pivot the data to create a 2D table with Sex on the y-axis and Pclass on the x-axis
pivot_table = data.pivot_table(index='sex', columns='class', values='survived', aggfunc='mean')

# Create the heatmap with different colors for each survival rate
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt='.2f', cbar=False, linewidths=.5,
            linecolor='gray', square=True, center=0.5)

sns
```
