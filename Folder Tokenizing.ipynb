## Ekonomi

```python
import os
from bs4 import BeautifulSoup

for file in os.listdir('C:/Users/ASUS/Downloads/berita/ekonomi'):
    with open("C:/Users/ASUS/Downloads/berita/ekonomi/{dir}".format(dir = file), encoding='utf-8') as beritaEkonomi:
        beritaEkonomi = BeautifulSoup(beritaEkonomi, 'html.parser')
        berita_Ekonomi = beritaEkonomi.get_text()
        
        # Case Folding
        # Lowercase
        lowercase = berita_Ekonomi.lower()
        
        import re # Regular Expression
        import string

        hasil = re.sub(r"\d+","",lowercase)
        hapus_simbol = hasil.translate(str.maketrans("","",string.punctuation))
        hapus_spasi = hapus_simbol.strip()
        
        # Menghapus Stopword
        from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
        factory = StopWordRemoverFactory()
        stopword = factory.create_stop_word_remover()
        stop = stopword.remove(hapus_spasi);

        # Tokenization
        import nltk
        from nltk.tokenize import word_tokenize
        tokens = nltk.tokenize.word_tokenize(stop)
    
        # Menghitung kemunculan kata di dalam 1 kalimat
        from nltk.probability import FreqDist
        frekuensi = nltk.FreqDist(tokens)
        jumlah_per_kata = str(frekuensi.most_common())
        
        print(jumlah_per_kata)
        
        # buka file untuk ditulis
        berita_Ekonomi_fix = open("C:/Users/ASUS/Downloads/berita/ekonomi/{nama}.txt".format(nama = file), "w", encoding='utf-8')

        # tulis teks ke file
        berita_Ekonomi_fix.write(jumlah_per_kata)

        # tutup file
        berita_Ekonomi_fix.close()
```

# tf idf



```python
def computeTF(wordDict, bagOfWords):
    tfDict = {}
    bagOfWordsCount = len(bagOfWords)
    for word, count in wordDict.items():
        tfDict[word] = count / float(bagOfWordsCount)
    return tfDict
```
